```python
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# 1. Data UMKM Desa (formal + informal)
# ==============================
data_all = [
    {"nama": "UMKM A", "sektor": "tempe", "kategori": "Pangan (koperasi)"},
    {"nama": "UMKM B", "sektor": "bakso", "kategori": "Pangan (koperasi)"},
    {"nama": "UMKM C", "sektor": "jamu", "kategori": "Pangan (pribadi)"},

    {"nama": "UMKM D", "sektor": "pop ice", "kategori": "Sektor lain"},
    {"nama": "UMKM E", "sektor": "pop ice", "kategori": "Sektor lain"},
    {"nama": "UMKM F", "sektor": "warung rumahan", "kategori": "Sektor lain"},
    {"nama": "UMKM G", "sektor": "abon lele ari des", "kategori": "Sektor lain"},
    {"nama": "UMKM H", "sektor": "anyaman", "kategori": "Kerajinan"},

    {"nama": "UMKM I", "sektor": "penjahit", "kategori": "Jasa"},
    {"nama": "UMKM J", "sektor": "penjahit", "kategori": "Jasa"},
]

df_all = pd.DataFrame(data_all)

# Hitung jumlah UMKM per sektor & kategori
jumlah_per_sektor_all = df_all.groupby(["kategori", "sektor"]).size().reset_index(name="jumlah")
jumlah_per_kategori_all = df_all.groupby("kategori").size().reset_index(name="total")

print("=== Jumlah UMKM per sektor (Semua) ===")
print(jumlah_per_sektor_all)
print("\n=== Jumlah UMKM per kategori (Semua) ===")
print(jumlah_per_kategori_all)

# ==============================
# Visualisasi Semua UMKM
# ==============================
colors = ["#2d5c75", "#ebc57e", "#48849f"]

# Bar chart jumlah UMKM per kategori
plt.figure(figsize=(6,4))
plt.bar(jumlah_per_kategori_all["kategori"], jumlah_per_kategori_all["total"],
        color=colors[:len(jumlah_per_kategori_all)])
plt.title("Total UMKM per Kategori (Semua)", fontsize=14)
plt.xlabel("Kategori")
plt.ylabel("Jumlah UMKM")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()

# Pie chart proporsi kategori
plt.figure(figsize=(6,6))
plt.pie(jumlah_per_kategori_all["total"], labels=jumlah_per_kategori_all["kategori"],
        autopct='%1.1f%%', startangle=90, colors=colors[:len(jumlah_per_kategori_all)])
plt.title("Proporsi UMKM per Kategori (Semua)", fontsize=14)
plt.show()


# ==============================
# 2. Data UMKM Koperasi Resmi Saja
# ==============================
data_koperasi = [
    {"nama": "UMKM A", "sektor": "tempe", "kategori": "Pangan (koperasi)"},
    {"nama": "UMKM B", "sektor": "bakso", "kategori": "Pangan (koperasi)"},
]

df_kop = pd.DataFrame(data_koperasi)

jumlah_per_sektor_kop = df_kop.groupby(["kategori", "sektor"]).size().reset_index(name="jumlah")
jumlah_per_kategori_kop = df_kop.groupby("kategori").size().reset_index(name="total")

print("\n=== Jumlah UMKM per sektor (Koperasi Resmi) ===")
print(jumlah_per_sektor_kop)
print("\n=== Jumlah UMKM per kategori (Koperasi Resmi) ===")
print(jumlah_per_kategori_kop)

# ==============================
# Visualisasi UMKM Koperasi Resmi
# ==============================

# Bar chart
plt.figure(figsize=(6,4))
plt.bar(jumlah_per_kategori_kop["kategori"], jumlah_per_kategori_kop["total"],
        color=colors[:len(jumlah_per_kategori_kop)])
plt.title("Total UMKM per Kategori (Koperasi Resmi)", fontsize=14)
plt.xlabel("Kategori")
plt.ylabel("Jumlah UMKM")
plt.xticks(rotation=20, ha="right")
plt.tight_layout()
plt.show()

# Pie chart
plt.figure(figsize=(5,5))
plt.pie(jumlah_per_kategori_kop["total"], labels=jumlah_per_kategori_kop["kategori"],
        autopct='%1.1f%%', startangle=90, colors=colors[:len(jumlah_per_kategori_kop)])
plt.title("Proporsi UMKM per Kategori (Koperasi Resmi)", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data fasilitas desa
fasilitas = [
    "Balai desa",
    "Kantor desa",
    "Aula desa",
    "Makam",
    "Sekolahan TK/PAUD/SD/SMP",
    "Masjid besar",
    "Mushola"
]

# Warna yang dipilih dari palet
colors = ["#ca983e", "#ddb263", "#ebc57e", "#f4d891",
          "#fbe9ac", "#194259", "#2d5c75", "#48849f"]

# --- Donut chart (biar estetik) ---
plt.figure(figsize=(7,7))
plt.pie([1]*len(fasilitas), labels=fasilitas, colors=colors, startangle=90,
        counterclock=False, wedgeprops=dict(width=0.4))
plt.title("Fasilitas Desa", fontsize=16, fontweight="bold")
plt.show()

# --- Bar chart estetik ---
plt.figure(figsize=(10,6))
bars = plt.bar(fasilitas, [1]*len(fasilitas), color=colors)
plt.title("Fasilitas Desa", fontsize=16, fontweight="bold")
plt.xticks(rotation=30, ha="right")
plt.yticks([])  # hapus angka jumlah
for bar, label in zip(bars, fasilitas):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height()/2, label,
             ha='center', va='center', rotation=90, fontsize=9, color="white", fontweight="bold")
plt.tight_layout()
plt.show()
```
